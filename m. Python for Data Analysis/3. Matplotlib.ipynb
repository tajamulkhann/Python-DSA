## **3. Subplots**

`subplots method provides a way to plot multiple plots on a single figure.`

* Use any command with the help of `set_`

### **Functional Subplot**  `Doubt`

`Creating Multiplots on Same Canvas`



````python
import numpy as np
import matplotlib.pyplot as plt

q = np.arange(0,10,1)
w = 2*q + 5
e = 3*q + 10

plt.subplot(1,2,1)   # 1 rows , 2 col, plot Number
plt.plot(q,w, 'b--*')
plt.title("graph1")

plt.subplot(1,2,2)    # 1 rows , 2 col, plot Number
plt.plot(q,e)
plt.title("graph2");
````

### **Object Oriented Subplots**  `(API Method)`

The main idea in using the more formal Object Oriented method is to create figure objects and then just call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it.

To begin we create a figure instance. Then we can add axes to that figure:

Yay.jpg

#### **Adding Custom Axes In Empty Canvas** `Method`

````python
fig1 = plt.figure()

custom_axes = fig1.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, width, height of axes
custom_axes.plot(q,w, 'b--*')
custom_axes.set_xlabel = 'yay';
````

**Understanding syntax of Add Axes**  `lbwh concept`

````python
fig2 = plt.figure(figsize=(5,5))
custom_axesa = fig2.add_axes([0.1,0.1,0.8,0.8]) #main Axes
custom_axesb = fig2.add_axes([0.2,0.5,0.4,0.3]) #Axes with in main axes
````

 Values always starts from 0 to 1
 ```
0.2 = Starts a 25% Left from 0 point
0.5 = Starts from bottom at 50 % height of canvas
0.4 = spreads to 50% of width of canvas
0.3 = spreads to 100% of length of canvas
```



**Plot in Added Axes**

---


````python
fig2 = plt.figure(figsize=(5,5))
custom_axes1 = fig2.add_axes([0.1,0.1,0.8,0.8]) #main Axes
custom_axes1.set_title('Big Plot')
custom_axes2 = fig2.add_axes([0.2,0.5,0.4,0.3])
custom_axes2.set_title('Small Plot')
custom_axes1.plot(q,'k--*')
custom_axes2.plot(e);
````

#### **Object Oriented Advanced Subplots**

Code is a little more complicated, but the advantage is that we now have full control of where the plot axes are placed, and we can easily add more than one axis to the figure:

* Syntax

```
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (n,n))

ax[0, 0].plot(x, 'c-o')
ax[0, 1].plot(y, 'g-*')
ax[1, 0].plot(z, 'k^')
ax[1, 1].plot(a, z, 'b-d')

ax[0, 0].set_title("Title")
ax[0, 1].set_title("Title")
ax[1, 0].set_title("Title")
ax[1, 1].set_title("Title");

```


**1. Simple Structure**  

`Same Like Adding Axes in empty Canvas`

````python
fig, ax = plt.subplots()
ax.plot(q,e);
````

**2. Advance Structure**  

`The Main Advantage is specifying as many nrows and ncols as we want`

````python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (15,7))
 #specify rows and cols as many rows

ax[0,0].plot(height, 'c-o')
ax[0, 1].plot(weight, 'g-*')
ax[1, 0].plot(age, 'k^')
ax[1, 1].plot(salary, age, 'b-d')

ax[0, 0].set_title("Dist of Height")
ax[0, 1].set_title("Dist of weight")
ax[1, 0].set_title("Dist of calories")
ax[1, 1].set_title("Dist of height, weight");
````

````python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (15,7))
 #specify rows and cols as many rows

ax[0].plot(height, age, color = 'blue')
ax[1].plot(weight,salary, 'g-*')

ax[1].set_title("Dist of calories")
ax[0].set_title("Dist of height, weight")

ax[0].legend(labels = ['height','age']);
````

**subplots()**
The plt.subplots() object will act as a more automatic axis manager.

Basic use cases:

````python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

# Now use the axes object to add stuff to plot
axes.plot(height, weight, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
````

A common issue with matplolib is overlapping subplots or figures. We ca use fig.tight_layout() or plt.tight_layout() method, which automatically adjusts the positions of the axes on the figure canvas so that there is no overlapping content:

````python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (15,7))

ax[0, 0].plot(height, 'c-o')
ax[0, 1].plot(weight, 'g-*')
ax[0, 0].legend(labels =['height'])
ax[1, 0].plot(age, 'k^')
ax[1, 1].plot(salary, age, 'b-d')

ax[0, 0].set_title("Dist of Height")
ax[0, 1].set_title("Dist of weight")
ax[1, 0].set_title("Dist of calories")
ax[1, 1].set_title("Dist of height, weight")

plt.tight_layout();
````

### **3. Scatter Plot** `Relationship`

Scatter plots are used to plot data points on a horizontal and a vertical axis To determine whether or not two variables have a relationship or correlation



````python
# Creating data
x=[10,20,30,40,50,60,70,80]
a=[2,3,4,9,7,5,6,1]
b=[9,7,6,5,4,3,2,3]

#create scatter plot
plt.scatter(x,a,color='m',alpha=0.5,edgecolors='k',marker='o',s=100)
plt.scatter(x,b,color='y',alpha=0.5,edgecolors='k',marker='D',s=100)
plt.legend([a,b])
plt.title('Scatter')
plt.xlabel('x')
plt.ylabel('a,b')
plt.grid()
plt.savefig("Scatter plot.png")
plt.show()
````

````python
plt.figure(figsize = (10, 5))
plt.scatter(tk['workclass'][:100], tk.education[:100], color = 'c', marker = '*', s = 50)
plt.grid();
````

### **4. Boxplot** `Distribution of Numerical Values`

A boxplot summarizes the distribution of a numeric variable for one or several groups. It allows to quickly get the median, quartiles and outliers but also hides the dataset individual data points

````python
# creating data
total = [10,20,3,40,50]
order = [20,30,60,4,80]
discount=[3,4,45,27,60]
data= list([total,order,discount])
plt.boxplot(data,showmeans=True)
plt.show()
````

### **5. Violinplot** `Attractive Version of Box Plot`

A violin plot is similar to  whisker or box plot do. As it shows several quantitative data across one or more categorical variables. It can be an effective and attractive way to show multiple data.

````python
# creating data
total = [10,20,3,40,50]
order = [20,30,60,4,80]
discount=[3,4,45,27,60]
data= list([total,order,discount])
plt.violinplot(data,showmeans=True)
plt.show()
````

````python
plt.figure(figsize = (10, 5))
plt.violinplot(tk['age'])
plt.grid();
````

### **6. Pie Chart** `represents Data in percents`

````python
fruits = {'mango':60,'apple':100,'banana':30,'orange':40}
x = list(fruits.keys())
y = list(fruits.values())
plt.pie(y,labels=x,autopct='%1.1f%%',startangle=90,shadow=True)
plt.show()
````
